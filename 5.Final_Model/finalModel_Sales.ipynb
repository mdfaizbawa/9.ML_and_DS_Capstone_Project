```python
import pandas as pd
from sklearn.model_selection import train_test_split 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier   
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
dataset=pd.read_csv('sales_prep.csv',index_col=None)


dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)


indep_X=dataset[['Holiday_Season', 'Customer_Rating', 'Average_Transaction_Value','Weekend_Sales_Percentage', 'Marketing_Channel_TV']]
dep_Y=dataset[['Sales_Revenue']]
```

```python
sd=StandardScaler()
indep_X=sd.fit_transform(indep_X)
sdy=StandardScaler()
dep_Y=sdy.fit_transform(dep_Y)
param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error'],'splitter':['best', 'random'],'max_features':['sqrt', 'log2']}
from sklearn.tree import DecisionTreeRegressor
grid=GridSearchCV(DecisionTreeRegressor(), param_grid, refit=True, verbose=3,n_jobs=-1)
grid.fit(indep_X,dep_Y)
```

```python
result=grid.cv_results_
print("Thte R_Score value for best parameter:{}".format(grid.best_params_))
table=pd.DataFrame.from_dict(result)
table
```

```python
y_predt=grid.predict(indep_X)
from sklearn.metrics import r2_score
r_score=r2_score(dep_Y,y_predt)
r_score
```

```python
filename="final_model_sales.sav"
pickle.dump(grid,open(filename,'wb'))
```
